# Load KIMORE K Fold

```python
import pickle  # Importing the pickle module for loading serialized data

# List of result filenames corresponding to different subsets of the dataset
result_list = [
    '1 JP all', '2 JO all', '3 JA all', 
    '4 JP healthy', '5 JO healthy', '6 JA healthy', 
    '7 JP disabled', '8 JO disabled', '9 JA disabled'
]

# File path to the directory where the k-fold split data files are stored
file_path = './k fold release kimore/'

# Iterating through each subset in the result list
for i in range(len(result_list)):
    # Constructing the full file path and opening the pickle file in read-binary mode
    with open(f'{file_path}{result_list[i]}.pkl', 'rb') as f:
        load_data = pickle.load(f)  # Loading the serialized data from the file

    # Printing the available keys in the loaded dictionary to understand its structure
    print('\n')
    print(load_data.keys())
    print(load_data['Kfold 2'].keys())
    
    # Printing the shape of the training and testing index arrays for the second k-fold split
    print(load_data['Kfold 2']['train index'].shape)  # Shape of the training set indices
    print(load_data['Kfold 2']['test index'].shape)   # Shape of the testing set indices

    # Displaying the first five training and testing indices for verification
    print(load_data['Kfold 2']['train index'][:5])
    print(load_data['Kfold 2']['test index'][:5])

```
